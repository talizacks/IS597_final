```python
c = pd.read_csv('clusters.csv')
c.rename(columns = {'geometry_x':'geometry'},inplace = True)
c['geometry_x'] = c['geometry'].apply(wkt.loads)
c.drop('geometry',inplace=True, axis=1)
c.rename(columns = {'geometry_y':'geometry'},inplace = True)
c['geometry_y'] = c['geometry'].apply(wkt.loads)
c.drop('geometry',inplace=True, axis=1)

c = gpd.GeoDataFrame(c, crs='epsg:4326',geometry = ('geometry_x'))
c.set_geometry('geometry_y',inplace=True)
c
```

```python
big_clusters = c.groupby(by = 'index_x').count().sort_values('Unnamed: 0',ascending = False).head(15)
big_clusters.rename(columns={'Unnamed: 0': 'Number of times 2 crashes occured within an hour/day'}, inplace=True)
```

```python
big_cluster_full = c.merge(big_clusters['Number of times 2 crashes occured within an hour/day'], how='inner',left_on='index_x',right_on = 'index_x')
```
